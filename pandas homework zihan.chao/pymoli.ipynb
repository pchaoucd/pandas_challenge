```python
n_ItemID = len(purchase_data['Item ID'].unique())
list_ItemID = purchase_data['Item ID'].unique()
list_ItemName = purchase_data['Item Name']
list_ItemPrice = purchase_data['Price']

new_data_ItemID = []

# creat a loop to calculate the purchase count
for i in range(n_ItemID):
    new_IDlist = []
    
    IDdata_i = purchase_data[purchase_data['Item ID']==list_ItemID[i]]
    ID_count = len(IDdata_i)
    
    Item_Name = list_ItemName[i]
    Item_Price = list_ItemPrice[i]



    purchase_value_sum = IDdata_i.Price.sum()
    
    new_IDlist.append(list_ItemID[i])
    new_IDlist.append(Item_Name)
    new_IDlist.append(ID_count)
    new_IDlist.append(Item_Price)
    new_IDlist.append(purchase_value_sum)
    
    new_data_ItemID.append(new_IDlist)
    
df_newID = pd.DataFrame(new_data_ItemID, columns = ['Item ID','Item Name','Purchase Count',
                                                    'Item Price','Total Purchase Value'])
df_newID = df_newID.set_index('Item ID')
df_new_ID = df_newID.sort_values(by='Purchase Count',ascending=False).head()

# add $ in front of the value 
                                          
# df_new_ID['Total Purchase Value'] = '$' + df_new_ID['Total Purchase Value'].astype(str)

df_new_ID
```
